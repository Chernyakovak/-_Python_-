## **Первичная обработка данных**


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Считываем файл CSV
df = pd.read_excel('/content/sample_data/База_данных.xlsx')

df.fillna(0, inplace=True)
df = df.replace('Мужской', 0)
df = df.replace('Женский', 1)

#print(df['Смотрите ли вы прямые эфиры матчей/соревнований?'].unique().tolist())
#['Смотрю всегда бесплатно', 'Смотрю иногда платно, иногда - бесплатно', 'Не смотрю', 'Смотрю всегда платно']

df = df.replace({'Смотрю всегда бесплатно': 1,
                 'Смотрю иногда платно, иногда - бесплатно': 2,
                 'Не смотрю': 3,
                 'Смотрю всегда платно': 4})

#print(df['Есть ли у вас дети?'].unique().tolist())
#['Да, 2 и более детей', 'Нет детей', 'Да, 1 ребенок']

# Заменяем значения в колонке с детьми
df = df.replace({'Нет детей': 0,
                 'Да, 1 ребенок': 1,
                 'Да, 2 и более детей': 2})

df['Город'] = df['Город'].str.replace("'", "", regex=False).str.lower().str.strip()
# Делаем первую колонку индексом
df = df.set_index(df.columns[0])
df.to_excel('data.xlsx')
df
```

```python
# 1. Настройка библиотек
import matplotlib.pyplot as plt
import seaborn as sns

# Настройка отображения графиков
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
# Для корректного отображения кириллицы на графиках
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python


# 2. Загрузка обработанной таблицы
df = pd.read_csv('data.csv', index_col=0)

# 3. Разделение столбцов на группы

# Первые 11 столбцов — хобби (общие занятия)
hobby_cols = df.columns[:11].tolist()

# Следующие 22 столбца (с 12 по 33) — виды спорта, за которыми следят (бинарные)
sport_interest_cols = df.columns[11:33].tolist()

# Столбцы вопросов о спорте (следующие 5, значения 0-4) — частота занятий/просмотра
sport_freq_cols = df.columns[33:38].tolist()

# Демографические столбцы (последние 4)
demo_cols = ['Пол', 'Возраст', 'Город', 'Есть ли у вас дети?']

print(f"\nКоличество хобби: {len(hobby_cols)}")
print(f"Виды спорта: {len(sport_interest_cols)}")
print(f"Вопросы о спорте: {sport_freq_cols}")
print(f"Демографические параметры: {demo_cols}")
```

```python

# 4. Анализ распределения демографических параметров

# 4.1 Пол (0 - мужчина, 1 - женщина)
plt.figure()
df['Пол'].value_counts().sort_index().plot(
    kind='bar', color=['#1f77b4', '#ff7f0e'], width = 0.5,
)
plt.title('Распределение по полу')
plt.xticks([0, 1], ['Мужчина', 'Женщина'], rotation=0)
plt.ylabel('Количество респондентов')
plt.show()

# 4.2 Возраст
plt.figure()
df['Возраст'].hist(bins=20, edgecolor='black', alpha=0.7, color='#2ca02c')
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()

# 4.3 Наличие детей (0 – нет, 1 – один, 2 – два и более)
plt.figure()
df['Есть ли у вас дети?'].value_counts().sort_index().plot(
    kind='bar', color='#9467bd'
)
plt.title('Наличие детей')
plt.xticks([0, 1, 2], ['Нет детей', 'Один', 'Два и более'], rotation=0)
plt.ylabel('Количество')
plt.show()

# 4.4 Топ 5 городов
top_cities = df['Город'].value_counts().head(5)
plt.figure()
top_cities.plot(kind='barh', color='#d62728')
plt.title('Топ 5 городов по числу респондентов')
plt.xlabel('Количество')
plt.gca().invert_yaxis()  # чтобы наибольший был сверху
plt.show()
```

```python

# 5. Общая популярность хобби
hobby_pop = df[hobby_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 10))
hobby_pop.plot(kind='barh', color='#17becf')
plt.title('Популярность хобби')
plt.xlabel('Количество')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# 6. Связь хобби с демографическими факторами

# 6.1 По полу
hobby_by_gender = df.groupby('Пол')[hobby_cols].mean()
hobby_by_gender.index = ['Мужчина', 'Женщина']

plt.figure(figsize=(14, 8))
hobby_by_gender.T.plot(kind='bar')
plt.title('Доля людей с хобби по полу')
plt.ylabel('Доля')
plt.xticks(rotation=30, ha='right')
plt.legend(title='Пол')
plt.tight_layout()
plt.show()
```

```python

# 6.2 По наличию детей
hobby_by_children = df.groupby('Есть ли у вас дети?')[hobby_cols].mean()
hobby_by_children.index = ['Нет детей', 'Один', 'Два и более']

plt.figure(figsize=(14, 10))
hobby_by_children.T.plot(kind='bar')
plt.title('Доля людей с хобби по наличию детей')
plt.ylabel('Доля')
plt.xticks(rotation=30, ha='right')
plt.legend(title='Дети')
plt.tight_layout()
plt.show()
```

```python

# 6.3 По возрастным группам
bins = [0, 25, 35, 45, 55, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56+']
df['Возраст_группа'] = pd.cut(df['Возраст'], bins=bins, labels=labels, right=False)

hobby_by_age = df.groupby('Возраст_группа')[hobby_cols].mean()

plt.figure(figsize=(14, 8))
hobby_by_age.T.plot(kind='bar')
plt.title('Доля людей с хобби по возрастным группам')
plt.ylabel('Доля')
plt.xticks(rotation=30, ha='right')
plt.legend(title='Возраст')
plt.tight_layout()
plt.show()
```

```python

# 6.4 По городам (топ-5)
top5_cities = top_cities.index.tolist()
df_top = df[df['Город'].isin(top5_cities)]
hobby_by_city = df_top.groupby('Город')[hobby_cols].mean()

plt.figure(figsize=(16, 8))
hobby_by_city.T.plot(kind='bar')
plt.title('Доля людей с хобби по топ 5 городам')
plt.ylabel('Доля')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Город', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python

# 7. Корреляционный анализ

# 7.1 Корреляция хобби с демографическими переменными
corr_data = df[hobby_cols + ['Пол', 'Возраст', 'Есть ли у вас дети?']].copy()
corr_matrix = corr_data.corr()
# Извлечём только корреляции хобби с демо-факторами
corr_with_demo = corr_matrix.loc[hobby_cols, ['Пол', 'Возраст', 'Есть ли у вас дети?']]

plt.figure(figsize=(10, 10))
sns.heatmap(corr_with_demo, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, cbar_kws={'label': 'Коэффициент корреляции'})
plt.title('Корреляция хобби с полом, возрастом и наличием детей')
plt.tight_layout()
plt.show()

#положительная корреляция с каким-либо хобби указывает на то, что у женщин это хобби встречается чаще (или среднее значение выше). Отрицательная — чаще у мужчин.


```

```python

# 7.2 Корреляция между самими хобби (кластеризация интересов)
plt.figure(figsize=(20, 16))
sns.heatmap(df[hobby_cols].corr(), cmap='viridis', center=0,
            linewidths=0.5, cbar_kws={'label': 'Корреляция'})
plt.title('Корреляционная матрица хобби')
plt.tight_layout()
plt.show()
```
